```python
#Import packages
import pandas as pd
import numpy as np
import os, sys
import pickle

#Import other files
raw_path = os.path.join('..', 'data', 'raw')
proc_path = os.path.join('..', 'data', 'processed')
sys.path.append(os.path.join('..', 'src'))
from model.Reg_Model import *
from model.Scoring import *

%load_ext autoreload
%autoreload 2
```

```python
#Read in data
# ss_v2 = pd.read_csv(os.path.join(proc_path, 'scoring_set_v2.csv'))

# Reads in data
seeds = pd.read_csv(os.path.join(raw_path, 'TourneySeeds_2017.csv'))
slots = pd.read_csv(os.path.join(raw_path, 'TourneySlots_2017.csv'))
results = pd.read_csv(os.path.join(raw_path, 'TourneyCompactResults_2017.csv'))
features = pd.read_csv(os.path.join(proc_path, 'team_features_v2_2017.csv'))

seeds = seeds[seeds['Season']==2017]
slots = slots[slots['Season']==2017]
results = results[results['Season']==2017]
```

```python
#Imports model
fn = os.path.join(proc_path, 'Models', 'model_v4.p')
model = pickle.load(open(fn, 'rb'))
```

```python
#Makes interactions
interactions = []
for idx1 in range(3):
    for idx2 in range(3):
        interactions.append(('clstr_'+str(idx1)+'_A', 'clstr_'+str(idx2)+'_B'))
interactions
```

```python
#Creates scorer instance to create SS
scorer = Scorer(features)
scorer.set_variables(slots, seeds, results, model)
# games = slots.rename(index=str, columns={'Strongseed':'Wteam', 'Weakseed':'Lteam'}).drop(['Round'], 1)
# games
pred = scorer.get_tourney_pred(interactions=interactions, just_pred=True)
```

```python
predictions = pd.DataFrame()
for p in pred:
    predictions = predictions.append(p)
predictions = predictions.reset_index(drop=True)
```

```python
predictions
```

```python
#merges with actual team names
teams = pd.read_csv(os.path.join(raw_path, 'Teams.csv'))

merged = predictions.merge(teams.rename(index=str, columns={'Team_Name':'Wteam_Name'}), left_on='Oteam', right_on='Team_Id')
merged = merged.drop(['Wteam', 'Lteam', 'matchup', 'Oteam', 'Team_Id'], 1)
```
