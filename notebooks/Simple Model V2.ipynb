```python
#Import packages
import pandas as pd
import numpy as np
import os, sys
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import pairwise_distances_argmin
from sklearn.cluster import AffinityPropagation

#Import other files
raw_path = os.path.join('..', 'data', 'raw')
proc_path = os.path.join('..', 'data', 'processed')
sys.path.append(os.path.join('..', 'src'))
from model.Reg_Model import *
%matplotlib inline  
```

```python
#Read in old scoring set data
ss_v1 = pd.read_csv(os.path.join(proc_path, 'scoring_set_v1.csv'))
```

```python
#Creates k-means clusters with 2, 3, 4, and 5 groups
x_cluster = ss_v1[['total_poss_W', 'oeff_W', 'deff_W']]
colors = ['#4EACC5', '#FF9C34', '#4E9A06', 'r', 'm']
fig = plt.figure()

for n_clusters in [2,3,4,5]:
    kmeans = KMeans(n_clusters = n_clusters, random_state = 0).fit(x_cluster)
    k_means_cluster_centers = np.sort(kmeans.cluster_centers_, axis=0)
    k_means_labels = pairwise_distances_argmin(x_cluster, k_means_cluster_centers)

    ax = fig.add_subplot(2, 2, n_clusters-1)
    for k, col in zip(range(n_clusters), colors):
        my_members = k_means_labels == k
        cluster_center = k_means_cluster_centers[k]
        ax.plot(x_cluster.ix[my_members, 0], x_cluster.ix[my_members, 1], 'w',
                markerfacecolor=col, marker='.')
        ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
                markeredgecolor='k', markersize=6)
    ax.set_title('KMeans: ' + str(n_clusters))
    #ax.set_xticks(())
    #ax.set_yticks(())
    
fig.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


from sklearn.cluster import KMeans

X = ss_v1[['oeff_W', 'deff_W', 'total_poss_W']]

y = ss_v1[['Outcome']]


estimators = {'k_means_2': KMeans(n_clusters=2, random_state = 0),
              'k_means_3': KMeans(n_clusters=3, random_state = 0),
              'k_means_4': KMeans(n_clusters=4, random_state = 0),
              'k_means_5': KMeans(n_clusters=5, random_state = 0)}


fignum = 1
for name, est in estimators.items():
    fig = plt.figure(fignum, figsize=(4, 3))
    plt.clf()
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

    plt.cla()
    est.fit(X)
    labels = est.labels_

    ax.scatter(X.ix[:, 0], X.ix[:, 1], X.ix[:, 2], c=labels.astype(np.float))

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('Offensive Eff')
    ax.set_ylabel('Defensive Eff')
    ax.set_zlabel('Possessions')
    fignum = fignum + 1
```

```python
#Creates Affinity-clusters with preference of -50
from itertools import cycle

x_cluster = ss_v1[['total_poss_W', 'oeff_W']]

af = AffinityPropagation(preference=-50).fit(x_cluster)
af_cluster_centers = np.sort(af.cluster_centers_, axis=0)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
n_clusters_ = len(cluster_centers_indices)

plt.figure(1)
plt.clf()

colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
for k, col in zip(range(n_clusters_), colors):
    class_members = labels == k
    cluster_center = x_cluster.ix[cluster_centers_indices[k]]
    plt.plot(x_cluster.ix[class_members, 0], x_cluster.ix[class_members, 1], col + '.')
    plt.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)
    #for x in x_cluster[class_members]:
    #    plt.plot([cluster_center[0], x[0]], [cluster_center[1], x[1]], col)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```

```python
#So arbitrarily, I'm going to use k-means clusters with 3 variables with 4 groups
ss_v2 = ss_v1.copy()

#Winning team
x_cluster_w = ss_v1[['oeff_W', 'deff_W', 'total_poss_W']]
cluster_model_w = KMeans(n_clusters=5, random_state = 0).fit(x_cluster_w)

clusters_w = pd.get_dummies(cluster_model_w.labels_)
idx = 0
for col in clusters_w.columns.values:
    col_name = 'clstr_w_' + str(idx)
    ss_v2[col_name] = clusters_w[col]
    idx += 1

#Losing team
x_cluster_l = ss_v1[['oeff_L', 'deff_L', 'total_poss_L']]
cluster_model_l = KMeans(n_clusters=5, random_state = 0).fit(x_cluster_l)

clusters_l = pd.get_dummies(cluster_model_l.labels_)
idx = 0
for col in clusters_l.columns.values:
    col_name = 'clstr_l_' + str(idx)
    ss_v2[col_name] = clusters_l[col]
    idx += 1
```

```python
#Exporting data to a csv called scoring set version 2
ss_v2.to_csv(os.path.join(proc_path, 'scoring_set_v2.csv'), index=False)
ss_v2 = pd.read_csv(os.path.join(proc_path, 'scoring_set_v2.csv'))
```
