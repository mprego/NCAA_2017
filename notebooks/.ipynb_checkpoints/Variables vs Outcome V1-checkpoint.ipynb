# Variables vs Tournament Outcomes
Looks at how previous variables compare with tournament outcomes

```python
#Import Packages
import pandas as pd
import numpy as np
import os, sys
import random

#Import other files
raw_path = os.path.join('..', 'data', 'raw')
proc_path = os.path.join('..', 'data', 'processed')
sys.path.append(os.path.join('..', 'src'))
from viz.Correl import *
from model.Scoring import *
%matplotlib inline  
%load_ext autoreload
%autoreload 2
```

```python
#Read in Kaggle data
results = pd.read_csv(os.path.join(raw_path, 'TourneyCompactResults.csv'))
teams = pd.read_csv(os.path.join(raw_path, 'Teams.csv'))

#Read in features data
features = pd.read_csv(os.path.join(proc_path, 'team_features.csv'))
```

```python
scorer = Scorer(features)
ss_v1 = scorer.make_ss(results)
ss_v1['Outcome'] = ss_v1['Wscore'] - ss_v1['Lscore']
ss_v1 = ss_v1[ss_v1['Season']>2003]
ss_v1 = ss_v1.drop(['Lscore', 'Wscore', 'Season'], 1)
```

```python
#Outputs outcome data
ss_v1.to_csv(os.path.join(proc_path, 'scoring_set_v1.csv'), index=False)
```
